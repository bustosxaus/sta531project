```python
hmm_pitches = []
```

```python
%%time
for i in range(3):
    model = hmm.MultinomialHMM(n_components=10,verbose=False,algorithm='viterbi',n_iter=5000)
    model.fit([clayton_cat])
    print('Iteration %d: %d' % (i,model.monitor_.iter))
    print('Score %d: %d' % (i,model.score([clayton_cat])))
    hmm_pitches.append(model.sample(50)[0])
```

```python
plt.rcParams['figure.figsize'] = (15,5)
plt.xlim(-5,55)
plt.ylim(-0.3, 3.3)
plt.yticks(np.arange(0, 4, 1), cdc)
plt.plot(hmm_pitches[0],'--bo')
plt.plot(hmm_pitches[1],'--gs')
plt.plot(hmm_pitches[2],'--rD')
plt.savefig('hhmsamples.pdf', dpi = 400, bbox_inches='tight')
pass
```
